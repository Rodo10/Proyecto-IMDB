## Regresion

```python
from scipy import stats
import pingouin
import statsmodels.formula.api as sm
```

```python
data = movies[['duration', 'director_facebook_likes', 'gross', 'budget', 'num_user_for_reviews', 'imdb_score', 'movie_facebook_likes']]
data
```

```python
data.corr()
```

```python
sns.pairplot(data)
```

```python
model_reg = 'movie_facebook_likes~budget'
lm1 = sm.ols(formula=model_reg, data=data).fit()
print(lm1.summary())
```

#### Test Beta

Beta es muy cercano a cero, por lo que es probable que el 'budget' de las peliculas no explique la cantidad de likes en Facebook. No esta influeciando el budget a la variable de interes.

#### R2
El ajuste es casi nulo, del 0.04%.

```python
model_reg2 = 'movie_facebook_likes~budget + duration + director_facebook_likes + gross + num_user_for_reviews + imdb_score'
lm2 = sm.ols(formula=model_reg2, data=data).fit()
print(lm2.summary())
```

#### Regression multiple

El ajuste del modelo dado por R2 no es bueno, no existe evidencia de una correlacion entre las variables de interes, por lo que no se recomienda utilizar este metodo para explicar el numero de likes de facebook
